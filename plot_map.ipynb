```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 10]

from env.static_env import TA_Static
from utils import load_train_config

row = 2
col = 6
config_name = "simple_het"

fig, ax = plt.subplots(row, col)

for r in range(row):
  for c in range(col):
    cfg = load_train_config(config_name)
    env = TA_Static(cfg)
    env.reset()

    task_info = env.task
    task_amount = env.task_total_num

    ax[r,c].set_title(config_name)
    ax[r,c].set_xlim((-0.1,1.1))
    ax[r,c].set_ylim((-0.1,1.1))
    ax[r,c].set_aspect("equal")

    task_colors = ["k","m","c"]
    for i, (t_type_name, t_info) in enumerate(task_info.items()):
        ax[r,c].scatter(task_info[t_type_name]["position"][0][:,0],task_info[t_type_name]["position"][0][:,1],
                    marker='x', s=25,  facecolor=task_colors[i], label=t_type_name)

    ax[r,c].legend()
plt.tight_layout()
plt.show()
```
